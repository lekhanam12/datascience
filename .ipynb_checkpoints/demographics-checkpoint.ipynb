```python
%matplotlib inline

import matplotlib.pyplot as plt
import geopandas as gpd
import palettable as pltt
import seaborn as sns
from seaborn import palplot
import pandas as pd
from pysal.lib import weights
import numpy as np

from sklearn import cluster
from sklearn.cluster import KMeans
```

### Data Loading

```python
# Data file paths
age_dta_path = 'data/Indicators/age_bins_district_t_pc11.dta'

# Read dta file in
age_df = pd.read_stata(age_dta_path)

indicators = ['pc11_state_id','pc11_district_id', 'age_65_t_share', 'age_70_t_share', 'age_75_t_share', 'age_80_t_share','age_85_t_share' ]

age_df = age_df[indicators]

# Display top of table
age_df.head()
```

```python
# General info
age_df.info()
```

```python
age_df.describe()
```

### Transforming the Data


```python
# Fill NaN value with 0
age_df = age_df.fillna(0.0)

# Change type
age_df.pc11_district_id = age_df.pc11_district_id.astype(int)

age_df = age_df.set_index('pc11_district_id')

# Display top of table
age_df.head()
```

```python
#Selecting indicators 
age_group = ['age_65_t_share', 'age_70_t_share', 'age_75_t_share', 'age_80_t_share','age_85_t_share' ]
```

```python
# Figure style
sns.set_style("dark")
# Create figure and axes (this time it's 15, arranged 5 by 3)
f, axs = plt.subplots(nrows=1, ncols=5, figsize=(30, 20))
# Make the axes accessible with single indexing
axs = axs.flatten()
# Start the loop over all the variables of interest
for i, col in enumerate(age_group):
    # select the axis where the map will go
    ax = axs[i]
    # Plot the map
    age_df.boxplot(column=col, showmeans=True, ax=ax)
# Display the figure
plt.show()
```

### Shape file
- Loading
- Dropping unnecessary columns
- Ploting the map

```python
# Loding the shape file
geo_map = gpd.read_file('data/ShapeFiles/2011_Dist.shp')

# Droping columns
geo_map = geo_map.drop([ 'DISTRICT', 'ST_NM', 'ST_CEN_CD', 'DT_CEN_CD'], axis=1 )

# Setup figure and axis
f, ax = plt.subplots(1, figsize=(8, 8))
# Add a layer with polygons on to axis `ax`
geo_map.plot(linewidth=0.5, ax=ax)
# Remove axis frame
ax.set_axis_off()
# Title
f.suptitle('District India PC 2011', size=15)
# Display
plt.show()

# Display top of table
geo_map.head()
```

### Joining the data
- Joing the shape file and the health infrastructure data
- Renaming the 'censuscode' column

```python
#Joining data 
age = geo_map.join(age_df, on = 'censuscode')

# renameing columns
age = age.rename(columns = {'censuscode' : 'pc11_district_id'})

# dropping all rows with missing data
age = age.dropna(axis=0)

# Display top of table
age.head()
```

```python
# General info
age.info()
```

```python
# Create figure and axes (this time it's 6, arranged 3 by 2)
f, axs = plt.subplots(nrows=3, ncols=2, figsize=(20, 40))
# Make the axes accessible with single indexing
axs = axs.flatten()
# Start the loop over all the variables of interest
for i, col in enumerate(age_group):
    # select the axis where the map will go
    ax = axs[i]
    # Plot the map
    age.plot(column=col, ax=ax, scheme='Quantiles',edgecolor='w', \
             linewidth=0.1, cmap='viridis_r', alpha=1, legend=True)
    # Remove axis clutter
    ax.set_axis_off()
    # Set the axis title to the name of variable being plotted
    ax.set_title(col)
# Display the figure
plt.show()
```

### Clustering


```python
# setting number of clusters
kmeans7 = cluster.KMeans(n_clusters=7)
kmeans7
```

***To run the algorithm the fit() method will be used***

```python
# This line is required to obtain the same results always. 
#(Computers are not truly random. They only seem that way. If this is interesting for you, ask me in class ;) Or try with a different seed.
np.random.seed(2000)
# Run the clustering algorithm
k7cls = kmeans7.fit(age[age_group])
```

```python
k7cls.labels_
```

***Each number is different category***

```python
age['k7cls'] = k7cls.labels_
```

***Evaluation of the number of the clusters will be made by using the Elbow Method - the elbow method serves to determine the optimal number of clusters.***

```python
# calculate distortion for a range of number of cluster
distortions = []
for i in range(1, 15):
    kmeans=KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    ) 
    kmeans.fit(age[age_group])
    distortions.append(kmeans.inertia_)

# plot
plt.plot(range(1, 15), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

***4 is the number where the line stop descending***
